## Population By Country Data (2023 - 2024)

```python
# Create an array of years from 2010 to 2024 (excluding 2025)
years = np.arange(2023, 2025)

# Initialize an empty list to store table rows
listofrows = []

# Iterate through each year in the 'years' array
for year in years:
    # Define the base URL for the population data by country
    base_url = "https://database.earth/population/by-country/"
    
    # Set the selected year for the current iteration
    year_selected = year
    
    # Construct the URL for the specific year's webpage
    webpage = base_url + str(year)
    
    # Send a GET request to retrieve the webpage content
    response = requests.get(webpage)
    
    # Create a BeautifulSoup object to parse the HTML content
    web_content = BeautifulSoup(response.text, "html.parser")
    
    # Find the HTML table containing population data
    table = web_content.find("table")
    
    # Find all rows (tr elements) within the table
    table_tr = table.find_all("tr")
    
    # Iterate through each row in the table (skipping the header row, assumed to be the first row)
    for row in table_tr[1:]:
        # Extract all cells (td elements) from the current row
        row_data = row.find_all("td")
        
        # Extract text from each cell and strip whitespace, then store in a list (each_row)
        each_row = [data.text.strip() for data in row_data]
        
        # Append the selected year to the end of each row's data
        each_row.append(year_selected)
        
        # Append the row (list) to the listofrows list
        listofrows.append(each_row)

# Create a pandas DataFrame from the listofrows with specified column names
df_ptwo = pd.DataFrame(listofrows, columns=["Country", "Population", "Change +/-", "Year"])

# Remove leading characters (assumed to be two characters) from the 'Country' column values
df_ptwo["Country"] = df_ptwo["Country"].str[2:]

# Select and reorder columns in the DataFrame
df_ptwo = df_ptwo[["Country", "Population", "Year"]]


#df_ptwo.to_csv("df_ptwo.csv")

df_ptwo.head()
```

## Continent, Country, Currency and Symbol Data

```python
# Define the URL from which to retrieve webpage content
url = "https://thefactfile.org/countries-currencies-symbols/"

# Send a GET request to the specified URL to retrieve webpage content
response_ok = requests.get(url)  # Use requests.get() to make a GET request to the URL

# The response_ok object contains the HTTP response returned by the server
# It includes the content of the webpage and metadata associated with the response
# You can inspect the response object to access the retrieved webpage content and other details
response_ok
```

```python
# Scrapping each row from the website table
content = BeautifulSoup(response_ok.text)
table_id = content.find("table")


# Find the table by its ID (if it has one)
table_id = content.find("table")

if table_id:
    # Find all rows in the table
    table_rows = table_id.find_all("tr")

    all_rows = []

    # Loop through each row (skipping the header row, assuming it's the first one)
    for row in table_rows[1:]:
        # Find all cells (td) in the row
        row_data = row.find_all("td")
        # Extract text from each cell and strip whitespace
        each_row = [data.text.strip() for data in row_data]

        # Find any footer cells (th) in the row
        foot_data = row.find_all("th")
        # Extract text from each footer cell and strip whitespace
        individual_row = [data.text.strip() for data in foot_data]

        # Append the footer data to the current row
        each_row.extend(individual_row)

        # Append the row to the list of all rows
        all_rows.append(each_row)

    # Now all_rows contains the data from the table, with footer data included where available
    print(all_rows)
else:
    print("Table with specified ID not found in the HTML.")
```

```python
# Create a DataFrame 'cont_curr' from 'all_rows' with specified column names
cont_curr = pd.DataFrame(all_rows, columns=["Continent", "Country", "Currency", "Code", "Symbol"])

# Display the shape (dimensions) of the DataFrame 'cont_curr'
print(cont_curr.shape) 

# Display the first few rows of the DataFrame 'cont_curr' for inspection
cont_curr.head()
```

```python
# Explore and Transform the Table

# Within this table is embeded the continent due to how it was structured from the website.

# Hence, I will need to split this table  into six parts, clean each part separately, then combine.


asia_one = cont_curr[:50]

asia_one["Continent"]  = "Asia"

asia_two = {
    "Continent": ["Asia", "Asia"],
    "Country": ["Hong Kong (SAR China)", "Macao (SAR China)"],
    "Currency": ["Hong Kong Dollar", "Pataca"],
    "Code": ["HKD", "MOP"],
    "Symbol": ["$", "MOP$"]
}

# Create DataFrame
df_hk = pd.DataFrame(asia_two)

asia = pd.concat([asia_one,df_hk])

print(asia.shape)

asia.head()
```
